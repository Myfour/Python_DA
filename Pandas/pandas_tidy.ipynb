```python
# 规整数据
import numpy as np
import pandas as pd
s1=pd.Series([100,200,300],index=['a','b','c'])
s2=pd.Series([400,500,600],index=['d','e','f'])
print(s1,'\n',s2)
```

```python
pd.concat([s1,s2])  # concat沿着轴向连接ndarray,未设置轴向，默认是0轴方向连接
```

```python
# DataFrame的连接
df1=pd.DataFrame([[110,120,130],[210,220,230],[310,320,330]])
df2=pd.DataFrame([[11,12,13],[21,22,23],[31,32,33]])
print(df1,'\n',df2)
```

```python
pd.concat([df1,df2],axis=1,join='inner')  # 索引按交集获取，上一个语句是用了默认的并集处理索引
```

```python
# 练习
pop=pd.read_csv('state-population.csv')
pop.head()
```

```python
areas=pd.read_csv('state-areas.csv')
areas.head()
```

```python
abbrevs=pd.read_csv('state-abbrevs.csv')
abbrevs.head()
```

```python
len(pop),len(abbrevs)
```

```python
pop_merged=pd.merge(pop,abbrevs,left_on='state/region',right_on='abbreviation',how='outer')
pop_merged
# 连接两张columns不同的表，分别选取两张表的某个字段来连接，left_on选坐表的一个字段，right_on选择右表的一个字段
# how参数控制连接的key的连接方式'left''right'其实就是类似于数据库中的left join 和 right join，就是以左表的所有键来，反之亦然
# 'outer' 是连接key的并集 'inner'是连接key的交集
```

```python
pop_merged=pop_merged.drop('abbreviation',axis=1)  # 删除1轴上的某一列
```

```python
pop_merged.head()
```

```python
# 检查规整后的数据是否有空
pop_merged.isnull().any()  # 检查结果是population和state中有空数据
```

```python
pop_merged[pop_merged['population'].isnull()].head()  # 查看population是NaN的记录
```

```python
pop_merged.loc[pop_merged['state'].isna(),'state/region'].unique()  # 发现state为NaN的所有记录的state/region只有PR和USA
```

```python
pop_merged.loc[pop_merged['state/region']=='PR','state']='Puerto Rico'
```

```python
pop_merged.loc[pop_merged['state/region']=='USA','state']='United States'
```

```python
pop_merged.isnull().any()
```

```python
pop_merged.head()
```

```python
areas.head()
```

```python
# 将areas的数据与pop_merged合并
result=pd.merge(pop_merged,areas,how='left')
result.head()
```

```python
result.isna().any()  # 查看数据缺失情况
```

```python
result.loc[result['area (sq. mi)'].isna(),'state'].unique()
```

```python
result.dropna(inplace=True)  # inplace属性实现原地修改，修改原始数据
result
```

```python
# 规整结束 查询数据
d2012=result.query('year==2012&ages=="total"')
d2012
```

```python
d2012.set_index('state',inplace=True)
d2012
```

```python
density=d2012['population']/d2012['area (sq. mi)']
density.head()
```

```python
density.sort_values(ascending=False,inplace=True)
density.head()
```

```python
density.tail()
```

```python
a=pd.Series([np.nan,2,4,np.nan,8,10])
b=pd.Series([1,2,np.nan,np.nan,5,np.nan])
```

```python
# pandas中提供了相同功能的函数
a.combine_first?  # 针对Series的话就是从a中选择数据，如果a中没有就从（other）传入的参数的Series中取
```

```python
a.combine_first(b)
```

```python
b.combine_first(a)
```
