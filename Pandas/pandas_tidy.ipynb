```python
import numpy as np
import pandas as pd
s1=pd.Series([100,200,300],index=['a','b','c'])
s2=pd.Series([400,500,600],index=['d','e','f'])
print(s1,'\n',s2)
```

```python
pd.concat([s1,s2])
```

```python
df1=pd.DataFrame([[110,120,130],[210,220,230],[310,320,330]])
df2=pd.DataFrame([[11,12,13],[21,22,23],[31,32,33]])
print(df1,'\n',df2)
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],keys=['df1','df2'])
```

```python
pd.concat([df1,df2],axis=1)  
```

```python
# 索引不再是位置索引的情况
df1.index=['a','c','d']
```

```python
df2.index=['b','c','d']
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=1,join='inner')  # 索引按交集获取，上一个语句是用了默认的并集处理索引
```

```python
df1.append(df2)
```

```python
df1.append(df2,ignore_index=True)
```

```python
# 合并数据
gdp1=pd.DataFrame({'city':['shanghai','guangzhou','shenzhen','chongqing'],'number':[27466.2,19610.9,19492.6,17558.8]})
gdp2=pd.DataFrame({'city':['shanghai','beijing','shenzhen','suzhou'],'number':[27466.2,24899.3,19492.6,15475.1]})
gdp1
```

```python
gdp2
```

```python
pd.merge(gdp1,gdp2)
```

```python
pd.concat([gdp1,gdp2],join='inner',axis=1)
```

```python
left_df=pd.DataFrame({'grade':['a','b','b','a','c','c','d'],'ldata':range(7)})
right_df=pd.DataFrame({'grade':['a','b','c'],'rdata':[30,40,50]})
left_df
```

```python
right_df
```

```python
pd.merge(left_df,right_df)
```

```python
pd.merge(left_df,right_df,on='grade',how='inner')
```
