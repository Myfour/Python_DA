```python
import numpy as np
```

```python
mu,sigma=0.0,1.0
g=np.random.normal(loc=mu,scale=sigma,size=1000)
g1=np.random.randn(1000)
```

```python
g=(g-g.min())/10
phy=np.round(((100.1-20)*g+20),1)
g1=(g1-g1.min())/10
mat=np.round(((100.1-20)*g1+20),1)
```

```python
mat
```

```python
np.mean(phy)
```

```python
np.std(phy)
```

```python
np.min(phy),np.max(phy)
```

```python
np.percentile(phy,80)
```

```python
np.median(phy)
```

```python
marks=np.vstack((phy,mat))
marks  # 生成1000名学生的物理与数学成绩
```

```python
marks.shape
```

```python
np.mean(marks,axis=1)  # 沿1轴计算平均值，得到物理与数学的平均值
```

```python
np.mean(marks,axis=0)  # 沿0轴计算每个人总成绩的平均值
```

```python
np.std(marks,axis=1)  # 计算1轴上的标准差
```

```python
#  获取物理成绩大于平均值的同学
phy[phy>np.mean(phy)]
```

```python
np.where(phy>mat,1,-1)  # 三元判断运算，第二个参数是为True时，第三个是为False时
```

```python
grand_a=np.mean(phy)+20
grand_b=np.mean(phy)
grand_a,grand_b
```

```python
# select 实现分段函数的功能，分段统计。
np.select([phy>=grand_a,np.logical_and(phy>=grand_b,phy<=grand_a)],['A','B'],default='C')
```
