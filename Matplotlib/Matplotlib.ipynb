```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib
```

```python
x=np.linspace(0,2*np.pi,100)
x
```

```python
y1=np.sin(x)
y2=np.cos(x)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
x=np.linspace(0,2*np.pi,100)
y=np.sin(x)
plt.plot(x,y)
y2=np.cos(x)
plt.plot(x,y2)
```

```python
fig=plt.figure()
```

```python
ax=fig.add_axes([0.1,0.1,0.8,0.8])
fig
```

```python
x=np.linspace(0,2*np.pi,100)
ax.plot(x,np.sin(x))
fig
```

```python
fig.savefig('sine.png')
```
